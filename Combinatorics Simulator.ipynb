```python
#Automatically test arbitrary numbers
#MADE BY ME

import random 
import numpy as np
from colorama import Fore, Back, Style

def singleGuess(printing=True, word='ABC', letters=None, circle=False, combination=False):
    if letters == None:
        letters = [*word]
        random.shuffle(letters)
    pastGuesses = []
    if printing:
        print(Fore.BLUE, 'Aiming to guess the word/phrase: "', word, '"', ' in any order,' if combination else (' in a circle ' if circle else '') , 'from the letters "', *letters, '"', Style.RESET_ALL, sep='')
    while True:
        passed = False
        newGuess = [*letters[:]]
        random.shuffle(newGuess)
        if combination:
            while sorted(newGuess[:len(word)]) in [sorted(i) for i in pastGuesses]:
                random.shuffle(newGuess)
            cutGuess = newGuess[:len(word)]
            cutGuess.sort()
            if cutGuess == sorted([*word]):
                passed = True
        else:
            while newGuess[:len(word)] in pastGuesses:
                random.shuffle(newGuess)
            cutGuess = newGuess[:len(word)]
            if circle:
                for i in range(len(word)):
                    if newGuess[i:]+newGuess[:i] == [*word]:
                        passed = True
            else:
                if cutGuess == [*word]:
                    passed = True
        if passed:
            if printing:
                print(Back.YELLOW, end='')
        if printing:
            print(*newGuess[:len(word)], Style.RESET_ALL, sep='', end=', ')
        if passed:
            if printing:
                print(Fore.BLUE, "\nDONE IN ", 1+len(pastGuesses), " GUESSES", Style.RESET_ALL, sep='')
            return 1+len(pastGuesses)
        else:
            pastGuesses += [cutGuess]
            
def multiGuess(n=10, printing=False, fin=True, word='ABC', letters=None, circle=False, combination=False):
    if letters == None:
        letters = [*word]
        random.shuffle(letters)
    else:
        letters = [*letters]
    counts = []
    for i in range(n):
        if printing:
            print(Fore.GREEN,"ATTEMPT ", i+1, " OF ", n, Style.RESET_ALL, sep='')
        counts += [singleGuess(printing=printing, word=word, letters=letters, circle=circle, combination=combination)]
    if fin:
        print(Fore.CYAN, '>>> ', 'The average of ', n, ' attempts to guess "', word, '" from "', *letters, '" randomly is ', sum(counts)/n, sep='')
    return counts

a = multiGuess(5,printing=True,fin=True,word='NO U', letters='NO UABC', circle=True, combination=True)
```
